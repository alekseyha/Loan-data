```python
# Load packages
import numpy as np 
import pandas as pd 
```

## Load your data

```python
# Load data from the csv file
df = pd.read_csv('loan_data.csv', index_col=None)

# Change the dots in the column names to underscores
df.columns = [c.replace(".", "_") for c in df.columns]
print(f"Number of rows/records: {df.shape[0]}")
print(f"Number of columns/variables: {df.shape[1]}")
df.head()
```

## Understand your variables

```python
# Understand your variables
variables = pd.DataFrame(columns=['Variable','Number of unique values','Values'])

for i, var in enumerate(df.columns):
    variables.loc[i] = [var, df[var].nunique(), df[var].unique().tolist()]
    
# Join with the variables dataframe
var_dict = pd.read_csv('variable_explanation.csv', index_col=0)
variables.set_index('Variable').join(var_dict)
```

```python
plt.figure(figsize=(20,12))
df[df['credit_policy']==1]['fico'].hist(alpha=0.7,color='green',
                                              bins=40,label='Credit_Policy=1')
df[df['credit_policy']==0]['fico'].hist(alpha=0.7,color='orange',
                                              bins=40,label='Credit_Policy=0')
plt.legend(fontsize=20)
plt.xlabel('FICO',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,12))
df[df['not_fully_paid']==1]['fico'].hist(alpha=0.7,color='purple',
                                              bins=40,label='not_fully_paid=1')
df[df['not_fully_paid']==0]['fico'].hist(alpha=0.7,color='pink',
                                              bins=40,label='not_fully_paid=0')
plt.legend(fontsize=20)
plt.xlabel('FICO', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='purpose',hue='not_fully_paid',data=df,palette='pastel')
plt.show()
```

```python
sns.jointplot(x='fico',y='int_rate',data=df, hue='not_fully_paid',height=15, s=100)
```

```python
sns.lmplot(y='int_rate',x='fico',data=df, hue='credit_policy',
           col='purpose', col_wrap=2, markers=["x", "o"], height=6)
```

```python
fig,axes = plt.subplots(3,2,figsize=(35,35))
for idx,cat_col in enumerate(['credit_policy', 'purpose', 'fico', 'inq_last_6mths', 'delinq_2yrs','pub_rec']):
    row,col = idx//2,idx%2
    sns.countplot(x=cat_col,data=df,hue='not_fully_paid',ax=axes[row,col], palette="Set3")
```

```python
fig,axes = plt.subplots(6,1,figsize=(10,80))
for idx,cat_col in enumerate(['int_rate', 'installment', 'log_annual_inc', 'dti', 'days_with_cr_line', 'revol_util']):
    sns.boxplot(y=cat_col,data= df,x='not_fully_paid',ax=axes[idx], dodge=False)

print(df[['int_rate', 'installment', 'log_annual_inc', 'dti', 'days_with_cr_line', 'revol_util']].describe())
```

```python
correlation = df[['int_rate', 'installment', 'log_annual_inc',
                'dti', 'fico', 'days_with_cr_line',
                'revol_bal', 'revol_util',
                'inq_last_6mths', 'delinq_2yrs']].corr(method='pearson')
fig = plt.figure(figsize = (9,7))
sns.heatmap(correlation, annot=True, cmap="YlGnBu", vmin=-1, vmax=+1)
plt.title('Correlation')
plt.show()
```
